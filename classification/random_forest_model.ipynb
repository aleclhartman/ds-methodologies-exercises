# Logistic Regression Walk-through Notes

- '.predict' vs '.predict_proba'
    - '.predict_proba' gives probabilities of classes; more flexible for estabilishing a threshold
    - '.predict' gives us classes (0 or 1); faster, but less control (uses .5 as the threshold)
- What does 'C' do? lower values decrease 
---
- A threshold is a value we choose
- If the probability of the model gives us is above the threshold, predict positive else negative
- When threshold is 0, everything will be positive (i.e., 1)
- When threshold is 1, everything will be negative (i.e., 0)
---
## Exploring the 'C' hyperparameter

Two things:

1. The 'C'hyperparameter: restrain size of coefficients
2. The python code

C is the inverse of the regularization

Takeaways:
1. Try experimenting with different values of C
2. Use a lower value of C to reduce overfitting
3. If C is important **scale your data**

```python
df = ac.get_titanic_data()
encoder, scaler, train, test = pr.prep_titanic(df)

# encode sex
encoder = LabelEncoder()
encoder.fit(train[["sex"]])
train.sex = encoder.transform(train[["sex"]])
test.sex = encoder.transform(test[["sex"]])

train, validate = train_test_split(train, random_state=74, train_size=.75)

X_train = train[["pclass", "sex", "age", "fare", "embarked", "alone"]]
X_validate = validate[["pclass", "sex", "age", "fare", "embarked", "alone"]]
X_test = test[["pclass", "sex", "age", "fare", "embarked", "alone"]]
y_train = train[["survived"]]
y_validate = validate[["survived"]]
y_test = test[["survived"]]
```

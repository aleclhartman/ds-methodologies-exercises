### 6. Is the sepal length significantly different in virginica compared to versicolor? Run an experiment to test this.

#### Hypotheses

$H_0$: The difference in sepal length between virginica and versicolor is insignificant

$H_a$: The difference in sepal length between virginica and versicolor is significant

```python
from scipy import stats
```

```python
alpha = .001

virginica = train[train.species == "virginica"].sepal_length
versicolor = train[train.species == "versicolor"].sepal_length

tstat, p = stats.ttest_ind(virginica, versicolor)
print(f"t = {tstat:.3f}")
print(f"p = {p}")

if p < alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

> We should hold onto the sepal_length feature for further analysis as the differences between the virginica and versicolor is significant.

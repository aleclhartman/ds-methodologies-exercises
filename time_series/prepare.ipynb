```python
def convert_date_dtype(df):
    """
    This function converts the Date dtype from an object to datetime
    """
    
    df["Date"] = pd.to_datetime(df.Date)
    
    return df
```

```python
def set_date_index(df):
    """
    This function sorts the Date field and sets it as the index of the DataFrame passed into the function
    """
    
    df = df.sort_values("Date").set_index("Date")
    
    return df
```

```python
germany = set_date_index(germany)
germany.head()
```

```python
def make_new_variables(df):
    """
    This function makes a month and year variable in the DataFrame passed into the function
    """
    
    df["month"] = df.index.month_name()
    df["year"] = df.index.year
    
    return df
```

## 1. Resample by the day and take the average temperature. Visualize the average temperature over time.

```python
temps = temps.sort_values("date").set_index("date")
temps
```

```python
avg_temps = pd.DataFrame(temps.temp.resample("D").mean())
avg_temps
```

```python
plt.figure(figsize=(16, 8))

plt.plot(avg_temps.temp, label="Mean")
plt.title("Temperatures Over Time")
plt.xlabel("Date")
plt.ylabel("Degrees")
plt.show()
```

---
## 2. Write the code necessary to visualize the minimum temperature over time.

```python
min_temps = pd.DataFrame(temps.temp.resample("D").min())
min_temps
```

```python
plt.figure(figsize=(16, 8))

plt.plot(avg_temps.temp, label="Mean Temps")
plt.plot(min_temps.temp, label="Min Temps")
plt.title("Temperatures Over Time")
plt.xlabel("Date")
plt.ylabel("Degrees")
plt.legend()
plt.show()
```

---
## 3. Write the code necessary to visualize the maximum temperature over time.

```python
max_temps = pd.DataFrame(temps.temp.resample("D").max())
max_temps
```

```python
plt.figure(figsize=(16, 8))

plt.plot(avg_temps.temp, label="Mean Temps")
plt.plot(min_temps.temp, label="Min Temps")
plt.plot(max_temps.temp, label="Max Temps")
plt.title("Temperatures Over Time")
plt.xlabel("Date")
plt.ylabel("Degrees")
plt.legend()
plt.show()
```

## 6. Resample by the day and calculate the min and max temp for the day (Hint: `.agg(['min', 'max'])`). Use this resampled dataframe to calculate the change in temperature for the day. Which month has the highest daily temperature variability?

```python
min_max_temps = temps.temp.resample("D").agg(["min", "max"])
min_max_temps
```

```python
min_max_temps["delta"] = min_max_temps["max"] - min_max_temps["min"]
min_max_temps
```

```python
min_max_mo_temps = pd.DataFrame(min_max_temps.delta.resample("M").max())
min_max_mo_temps["month"] = min_max_mo_temps.index.month_name()
min_max_mo_temps
```

```python
min_max_mo_temps.sort_values("delta", ascending=False).head(1)["month"]
```

## 1. Convert any negative delays to 0.

```python
flights = data.flights_20k()
flights
```

```python
flights["delay"] = np.where(flights.delay < 0, 0, flights.delay)
flights
```

---
## 2. Which hour of the day has the highest average delay?

```python
flights = flights.set_index("date")
flights
```

```python
mean_hourly_delays = pd.DataFrame(flights.delay.resample("H").mean())
mean_hourly_delays["hour"] = mean_hourly_delays.index.hour
mean_hourly_delays
```

```python
mean_hourly_delays.groupby("hour").delay.max().sort_values(ascending=False).head(1)
```

```python
flights.groupby("weekday").delay.sum().sort_values(ascending=False)
```

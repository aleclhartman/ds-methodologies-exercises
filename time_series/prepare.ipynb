## 5. Make sure all the work that you have done above is reproducible. That is, you should put the code above into separate functions and be able to re-run the functions and get the same results.

```python
germany = ac.get_germany()
germany.info()
```

```python
def convert_date_dtype(df):
    """
    This function converts the Date dtype from an object to datetime
    """
    
    df["Date"] = pd.to_datetime(df.Date)
    
    return df
```

```python
germany = convert_date_dtype(germany)
germany.info()
```

```python
def plot_dist(df):
    """
    This function plots the distributions of each variable from the DataFrame passed into the function
    """
    
    df.hist(figsize=(16, 16))
    plt.show()
```

```python
plot_dist(germany)
```

```python
def set_date_index(df):
    """
    This function sorts the Date field and sets it as the index of the DataFrame passed into the function
    """
    
    df = df.sort_values("Date").set_index("Date")
    
    return df
```

```python
germany = set_date_index(germany)
germany.head()
```

```python
def make_new_variables(df):
    """
    This function makes a month and year variable in the DataFrame passed into the function
    """
    
    df["month"] = df.index.month_name()
    df["year"] = df.index.year
    
    return df
```

```python
germany = make_new_variables(germany)
germany
```
